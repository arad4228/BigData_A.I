# Pandas 데이터 편집

### Pandas 데이터 편집
**concatenation**

```python
df1=pd.DataFrame (
    {
        "A":['a00','a01','a02'],
        "B":['b00','b01','b02'],
        "C":['c00','c01','c02']
    },
    index = ['a','b','c']
)
```

```python
df1
```

```python
df2=pd.DataFrame(
    {
        "B":['B00','B01','B02'],
        "C":['C00','C01','C02'],
        "D":['D00','D01','D02']
    },
    index = ['b','c','d']
)
```

```python
df2
```

```python
# Concatenation
pd.concat([df1, df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0, join="inner")
```

```python
pd.concat([df1, df2], axis=0,join="outer")
```

```python
pd.concat([df1, df2], axis=1, join="inner")
```

```python
pd.concat([df1, df2], axis=1,join="outer")
```

### Pandas 데이터 편집
**merge**

```python
df1 = pd.DataFrame(
    {
        "key":["k0", "k1", "k2", "k3"],
        "A":["a00", "a01", "a02", "a03"],
        "B":["b00", "b01", "b02", "b03"]
    }
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
        "key":["k1", "k2", "k3", "k4"],
        "C":["c01", "c02", "c03", "c04"],
        "D":["b01", "b02", "b03", "b04"]
    }
)
```

```python
df2
```

```python
pd.merge(df1, df2, on="key")
```

```python
pd.merge(df2,df1, on="key")
```

```python
df3 = pd.DataFrame(
    {
        "key1":["k0", "k1", "k2", "k3"],
        "key2":["k1","k2","k3","k3"],
        "A":["a00", "a01", "a02", "a03"],
        "B":["a10", "a11", "a12", "a13"]
    }
)
```

```python
df3
```

```python
df4 = pd.DataFrame(
    {
        "key1":["k0", "k1", "k1", "k3"],
        "key2":["k1","k2", "k2","k4"],
        "A":["b00", "b01", "b02", "b03"],
        "B":["b10", "b11", "b12", "b13"]
    }
)
```

```python
df4
```

```python
pd.merge(df3, df4, on="key1")
```

```python
pd.merge(df3,df4, on=["key2"])
```

```python
pd.merge(df3, df4, on=["key1","key2"])
```

```python
pd.merge(df3, df4, how="left", on=["key1", "key2"])
```

```python
pd.merge(df3, df4, how="right", on=["key1", "key2"])
```
