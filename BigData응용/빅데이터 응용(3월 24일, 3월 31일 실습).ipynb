# Regular Expression을 이용한 Scraping
### findall: 조건을 만족시키는 문자열을 찾아라

```python
str = "니가 185번 버스를 타고 가다가 130번으로 갈아타면 될 거 같은데, 비용은 1,500원이 들거야"
re.findall('[\d]+',str)
```

### finditer: 조건을 만족하는 문자열을 찾아 iter형으로 저장

```python
str = '우리가 만약 Tom과 만나게 도ㅚ면 jane은 뭐라고 할까? 울리엄은? 그러면 Edward은? 그럼 K는?'
re_iter = re.finditer('[A-Z][a-z]+', str)
print(type(re_iter))
for s in re_iter:
  print(s)
```

### match: 첫번째 문자열에 해당되는 문자열이 두번째 문자열에 있는지

### fullmatch: 첫번째 문자열에 완전히 일치하는 문자열이 두번째 문자열에 있는지

```python
print(re.match("Tom","Tom"))
print(re.match("Tom","TomTom"))
print(re.fullmatch("Tom","Tom"))
print(re.fullmatch("Tom","TomTom"))
```

### split: 첫번째 문자열로 두번째 문자열을 분할하라

```python
str1 = "강"
str2 = "우리나라의 4대 방어선은 압록강, 청천강, 두만강, 임진강이다"

s = re.split(str1,str2)
for t in s:
  # 앞의 4개의 list와 마지막 list의 차이는 원소의 수 이다.
  tt = re.split(' ', t)
  #print(tt)
  if(len(tt) > 1):
    print(tt[-1]+"강")
```

### sub: 세번째 문자열에 나오는 첫번째 문자열을 두번째 문자열로 교체하라

```python
str1 = "Jane"
str2 = "Mary"
str3 = "how can I say to you, Jane"
re.sub(str1, str2, str3)
```

```python
urls = []
titles = []
authors = []
prices = []
dates = []
brands = []

for item in re.findall(r'<td.*?</td?>', text, re.DOTALL):
  if re.findall(r'<td class="left">.*?</td>', item, re.DOTALL):
    res = re.search(r'<a href="(.*?)">', item)
    if ( res != None):
      url = re.search(r'<a href="(.*?)">', item).group(1)
      url = "https://www.hanbit.co.kr" + url
      urls.append(url)
      title = re.sub(r'<.*?>',' ',item)
      titles.append(title)
    else :
      author = re.sub(r'<.*?>', ' ', item)
      authors.append(author)
  # date
  if (re.findall(r'<td>.*?/td>', item, re.DOTALL)):
    date = re.sub(r'<.*?>',' ', item)
    dates.append(date)
  # price
  if (re.findall(r'<td class="right">.*?</td>', item, re.DOTALL)):
    price = re.sub(r'<.*?>', ' ', item)
    prices.append(price)
  # brand
  if (re.findall(r'<td class="brd_.*">.*?</td>', item, re.DOTALL)):
    brand = re.sub(r'<.*?>',' ', item)
    brands.append(brand)
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "URL": urls,
        "TITLE": titles,
        "AUTHOR": authors,
        "DATE": dates,
        "PRICE": prices,
        "BRAND": brands
    }
)
```

### find(tag) / select_one (tag): tag로 시작하는 끝나는 첫번쨰 문장을 찾아준다.

```python
soup = bs(text,"html.parser")
result = soup.find("tr")
print(result)
```

### find_all(tag) / select(tag): tag로 시작하는 모든 문장을 찾아준다.

```python
result = soup.find_all("tr")
print(result[0])
print("\n")
print(result[1])
type(result)
```

### string / text(): 문장내에서 tag로 묶이지 않은 문자열을 찾아준다.

```python
result = soup.select("tr")
print(result[1])
for t in result[2]:
  print(t)
  print(t.string)
```

```python
url2 = []
title2 = []
author2 = []
date2 = []
brand2 = []
price2 = []

for item in result:
  it1 = item.find_all("a")
  it2 = item.find_all("td")
  for i1 in it1:
    url = i1.get("href")
    if (url == None):
      continue
    url = "http://www.hanbit.co.kr" + url
    url2.append(url)
    brand2.append(it2[0].string)
    title2.append(it2[1].string)
    author2.append(it2[2].string)
    date2.append(it2[3].string)
    price2.append(it2[4].string)
```

```python
import pandas as pd
df2 = pd.DataFrame(
    {
        "URL": url2,
        "TITLE": title2,
        "AUTHOR": author2,
        "DATE": date2,
        "PRICE": price2,
        "BRAND": brand2,
    }
)
```

```python
import requests
crawling_url = 'http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?orderClick=d79'
response = requests.get(crawling_url)
text = response.text
```

```python
bestseller = soup.find('ul',{'class':'list_type01'})
bestseller
```

```python
bestseller_list = bestseller.findAll('div',{'class':'detail'})
len(bestseller_list)
```

```python
title3_list = []
for b_detail in bestseller_list:
  temp = b_detail.find('div',{'class':'title'})
  title = temp.find('strong').text
  title3_list.append(title)

print(title3_list)
```

```python
for b_detail in bestseller_list:
  title = b_detail.find('div',{'class':'title'}).find('strong').text
  title3_list.append(title)
```

```python
import re

def no_space(text):
  text1 = re.sub('\nbsp; | &nbsp; | \n|\t|\r','',text)
  text2 = re.sub('\n\n', '', text1)
  return text2
```

```python
import re

author_list = []
publisher_list = []
date_list = []
price_list = []

def get_item(text, idx):
  text2 = ""
  cnt = 0
  for t in text:
    if( (len(text2) == 0) & (t == " ")):
      continue
    if (t == "\n"):
      continue
    if (cnt == idx-1):
      if(t != "|" ):
        text2 = text2 + t
    if (t == '|'):
      cnt +=1
    if (cnt == idx):
      break
  return text2
  

for b_detail in bestseller_list:
  temp = b_detail.find('div',{'class':'author'})
  temp2 = temp.get_text()
  output = re.sub(r'저자', '', temp2)
  output = re.sub(r'더보기', '',output)

  output = no_space(output)
  
  author = get_item(output,1)
  author_list.append(author)

  publisher = get_item(output,2)
  publisher_list.append(publisher)

  date = get_item(output,3)
  date_list.append(date)

  temp3 = b_detail.find('div',{'class':'price'})
  price = temp3.find('strong').text
  price_list.append(price)
```

```python
print(len(title3_list))
print(len(author_list))
print(len(date_list))
print(len(price_list))
print(len(publisher_list))
```

```python
import pandas as pd
df3 = pd.DataFrame (
    {
        "TITLE": title3_list,
        "AUTHOR": author_list,
        "DATE": date_list,
        "PRICE": price_list,
        "PUBLISHER": publisher_list
    }
)
```
